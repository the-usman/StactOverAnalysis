```python
def multi_option_breaker (series) :
    df = series.to_frame()
    options = []
    for index, value in series[series.notnull()].items() :
        for option in value.split(';') :
            if not option in df.columns : 
                options.append(option)
                df[option] = False
            df.at[index, option] = True
    return df[options]
```
